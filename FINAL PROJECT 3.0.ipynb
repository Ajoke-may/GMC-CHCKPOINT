```python
import tkinter as tk
from tkinter import messagebox
import random

# Function to determine the winner
def play(player_choice):
    global player_score, computer_score

    options = ['Rock', 'Paper', 'Scissors']
    computer_choice = random.choice(options)
    
    if player_choice == computer_choice:
        result.set(f"Draw! Computer chose {computer_choice}.")
    elif (player_choice == 'Rock' and computer_choice == 'Scissors') or \
         (player_choice == 'Scissors' and computer_choice == 'Paper') or \
         (player_choice == 'Paper' and computer_choice == 'Rock'):
        result.set(f"You Win! Computer chose {computer_choice}.")
        player_score += 1
    else:
        result.set(f"You Lose! Computer chose {computer_choice}.")
        computer_score += 1

    player_score_label.config(text=f"Player Score: {player_score}")
    computer_score_label.config(text=f"Computer Score: {computer_score}")

# Function to reset the game
def Reset():
    global player_score, computer_score
    player_score = 0
    computer_score = 0
    player_score_label.config(text="Player Score: 0")
    computer_score_label.config(text="Computer Score: 0")
    result.set("Make your choice!")

# Function to exit the game
def Exit():
    root.destroy()

# Initialize the main window
root = tk.Tk()
root.title("Rock-Paper-Scissors")
root.geometry("400x400")
root.configure(bg='orange')

# Variables to store scores
player_score = 0
computer_score = 0

# Variable to store the result
result = tk.StringVar()
result.set("Make your choice by clicking the button!")

# Create the GUI elements
tk.Label(root, text='Rock, Paper, Scissors Game', font='"Arial Black" 18 bold', bg='orange').pack(pady=20)

tk.Button(root, text="Rock", width=15, command=lambda: play('Rock')).pack(pady=5)
tk.Button(root, text="Paper", width=15, command=lambda: play('Paper')).pack(pady=5)
tk.Button(root, text="Scissors", width=15, command=lambda: play('Scissors')).pack(pady=5)

tk.Label(root, textvariable=result, font=("Arial Black", 12), bg='orange').pack(pady=20)

player_score_label = tk.Label(root, text="Player Score: 0", font=("Arial Black", 12), bg='orange')
player_score_label.pack(pady=5)
computer_score_label = tk.Label(root, text="Computer Score: 0", font=("Arial Black", 12), bg='orange')
computer_score_label.pack(pady=5)

tk.Button(root, text="Reset", width=15, command=Reset).pack(pady=5)
tk.Button(root, text="Exit", width=15, command=Exit).pack(pady=10)

# Start the main event loop
root.mainloop()
```
