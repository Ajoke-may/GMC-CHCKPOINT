```python
import tkinter as tk
from tkinter import messagebox
import random

# Function to determine the winner
def play():
    player_choice = player_choice_var.get()
    options = ['Rock', 'Paper', 'Scissors']
    computer_choice = random.choice(options)
    
    if player_choice == computer_choice:
        result.set(f"Draw! Computer chose {computer_choice}.")
    elif (player_choice == 'Rock' and computer_choice == 'Scissors') or \
         (player_choice == 'Scissors' and computer_choice == 'Paper') or \
         (player_choice == 'Paper' and computer_choice == 'Rock'):
        result.set(f"You Win! Computer chose {computer_choice}.")
        global player_score
        player_score += 1
    else:
        result.set(f"You Lose! Computer chose {computer_choice}.")
        global computer_score
        computer_score += 1

    player_score_label.config(text=f"Player Score: {player_score}")
    computer_score_label.config(text=f"Computer Score: {computer_score}")

# Function to reset the game
def Reset():
    global player_score, computer_score
    player_score = 0
    computer_score = 0
    player_score_label.config(text="Player Score: 0")
    computer_score_label.config(text="Computer Score: 0")
    result.set("Make your choice!")

# Function to exit the game
def Exit():
    root.destroy()

# Initialize the main window
root = tk.Tk()
root.title("Rock-Paper-Scissors")
root.geometry("400x300")
root.configure(bg='orange')

# Variables to store scores
player_score = 0
computer_score = 0

# Variable to store the result
result = tk.StringVar()
result.set("Make your choice by clicking!")

# Variable to store player's choice
player_choice_var = tk.StringVar()

# Create the GUI elements
tk.Label(root, text='Rock, Paper, Scissors Game', font='"Arial Black" 12 bold', bg='orange').pack(pady=20)

tk.Radiobutton(root, text="Rock", variable=player_choice_var, value="Rock", bg='orange').pack(pady=5)
tk.Radiobutton(root, text="Paper", variable=player_choice_var, value="Paper", bg='orange').pack(pady=5)
tk.Radiobutton(root, text="Scissors", variable=player_choice_var, value="Scissors", bg='orange').pack(pady=5)

tk.Entry(root, textvariable=result, font=("Arial Black", 15), bg='orange', state='readonly', width=30).pack(pady=20)

player_score_label = tk.Label(root, text="Player Score: 0", font=("Arial Black", 15), bg='orange')
player_score_label.pack(pady=5)
computer_score_label = tk.Label(root, text="Computer Score: 0", font=("Arial Black", 15), bg='orange')
computer_score_label.pack(pady=5)

tk.Button(root, text="PLAY", width=15, command=play).pack(pady=5)
tk.Button(root, text="RESET", width=15, command=Reset).pack(pady=5)
tk.Button(root, text="EXIT", width=15, command=Exit).pack(pady=10)

# Start the main event loop
root.mainloop()
```
